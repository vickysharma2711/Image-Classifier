```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
Y = iris.target
```

```python
import tensorflow
from keras.utils import to_categorical

Y = to_categorical(Y)
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train, Y_test = train_test_split(X , Y , random_state=42 , test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

cnn = Sequential()

cnn.add(Dense(64 , activation='relu' , input_shape = (X_train.shape[1],)))
cnn.add(Dense(32 , activation='relu')),
cnn.add(Dense(16 , activation='relu')),
cnn.add(Dense(3 , activation='softmax'))
```

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(X_train , Y_train , batch_size=16 , epochs=150 , validation_split=0.2, verbose=1)
```

```python
loss , accuracy = cnn.evaluate(X_test , Y_test , verbose=0)
print(f"\n Test accuracy: {accuracy*100:.2f}%")
```

```python
prediction = cnn.predict(X_test)
print("\n🔹 Sample Predictions:")
print("Predicted:", prediction[0].argmax(), "| Actual:", Y_test[0].argmax())
```
